# Preprocessing

```python
# Fetch dataset
mushrooms = pd.read_csv("mushrooms.csv")

# Convert categorical data to numeric dummy variables
mushrooms = pd.get_dummies(mushrooms)

y = mushrooms['class_p']
X = mushrooms[[i for i in mushrooms.columns if i not in ['class_e', 'class_p']]]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Reshape Data for CNN

```python
X_train_cnn = X_train.to_numpy().reshape(X_train.shape[0], X_train.shape[1], 1)
X_test_cnn = X_test.to_numpy().reshape(X_test.shape[0], X_test.shape[1], 1)
```

# Create Neural Network classifier with TensorFlow

```python
# CNN classifier
nn_clf = Sequential([
    Input(shape=(X_train_cnn.shape[1], 1)),
    Conv1D(filters=32, kernel_size=3, activation='relu'),
    MaxPooling1D(pool_size=2),
    Flatten(),
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid')
])

nn_clf.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

history = nn_clf.fit(X_train_cnn, y_train, epochs=10, batch_size=32, verbose=1)

# Apply trained Neural Network on X test dataset to make predictions for y
y_pred_nn_probs = nn_clf.predict(X_test_cnn)
y_pred_nn = (y_pred_nn_probs > 0.5).astype(int).flatten()
y_pred_nn_bool = y_pred_nn.astype(bool)
```

# View Results

```python
# Predictions
print("Prediction:", y_pred_nn_bool)
print("\nNeural Network Test Metrics:")

y_train_pred_probs = nn_clf.predict(X_train_cnn)
y_train_pred = (y_train_pred_probs > 0.5).astype(int).flatten()

y_test_pred_probs = nn_clf.predict(X_test_cnn)
y_test_pred = (y_test_pred_probs > 0.5).astype(int).flatten()

print("Training accuracy:", accuracy_score(y_train, y_train_pred))
print("Test accuracy:", accuracy_score(y_test, y_test_pred))

print("Precision:", precision_score(y_test, y_test_pred, pos_label=1))
print("Recall:", recall_score(y_test, y_test_pred, pos_label=1))
print("F1 Score:", f1_score(y_test, y_test_pred, pos_label=1))
```

# Imports

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
```

# Preprocessing

```python
# Fetch dataset 
mushrooms = pd.read_csv("mushrooms.csv")

# https://stackoverflow.com/questions/75997811/how-to-get-rid-of-error-could-not-convert-string-to-float
mushrooms = pd.get_dummies(mushrooms)

y = mushrooms['class_p']  
X = mushrooms[[i for i in mushrooms.columns if i not in ['class_e', 'class_p']]]

# https://www.geeksforgeeks.org/machine-learning/sklearn-perceptron/
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Create SVM classifier with scikit-learn

```python
# SVM classifier
svm_clf = SVC(kernel='linear', random_state=42)
svm_clf.fit(X_train, y_train)

# Apply trained SVM on X test dataset to make predictions for y
y_pred_svm = svm_clf.predict(X_test)
```

# View Results

```python
# Predictions
print("Prediction:", y_pred_svm)

print("\nSVM Test Metrics:")

y_train_pred = svm_clf.predict(X_train)
y_test_pred = svm_clf.predict(X_test)

print("Training accuracy:", accuracy_score(y_train, y_train_pred))
print("Test accuracy:", accuracy_score(y_test, y_test_pred))

print("Precision:", precision_score(y_test, y_test_pred, pos_label=1))
print("Recall:", recall_score(y_test, y_test_pred, pos_label=1))
print("F1 Score:", f1_score(y_test, y_test_pred, pos_label=1))
```

```python
# Install the ucimlrepo package 
# pip install ucimlrepo

import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import pandas as pd
```

```python
# Import the dataset into your code 
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
mushroom = fetch_ucirepo(id=73) 
  
# data (as pandas dataframes) 
X = mushroom.data.features 
y = mushroom.data.targets 

# https://stackoverflow.com/questions/75997811/how-to-get-rid-of-error-could-not-convert-string-to-float
X = pd.get_dummies(X)
y = y.values.ravel()
  
# metadata 
# print(mushroom.metadata) 
  
# variable information 
# print(mushroom.variables)

# https://www.geeksforgeeks.org/machine-learning/sklearn-perceptron/
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Create SVM classifier
svm_clf = SVC(kernel='linear', random_state=42)
svm_clf.fit(X_train, y_train)

# Apply trained SVM on X test dataset to make predictions for y
y_pred_svm = svm_clf.predict(X_test)

# Print Prediction
print("SVM prediction: ", y_pred_svm)

# Print the accuracy_score of the model 
print("SVM Training accuracy:", svm_clf.score(X_train, y_train))
print("SVM Test accuracy:", svm_clf.score(X_test, y_test))
```
